```python
import random
import networkx as nx
import matplotlib.pyplot as plt

# Number of topics
N = 250  # You can adjust this between 200-300

# Generate a list of topic names
topics = [f"Topic {i}" for i in range(1, N+1)]

# Initialize the graph
G = nx.DiGraph()

# Add the root node
G.add_node(topics[0])

# Build the tree
for i in range(1, N):
    # Each topic will have exactly one prerequisite to maintain a tree structure
    parent = random.randint(0, i-1)
    G.add_edge(topics[parent], topics[i])

# Function to perform topological sort (sequence of studying topics)
study_order = list(nx.topological_sort(G))

# Output the study order
print("Recommended Study Order:")
for idx, topic in enumerate(study_order):
    print(f"{idx+1}. {topic}")

# Draw the graph
plt.figure(figsize=(15, 10))
pos = nx.nx_agraph.graphviz_layout(G, prog='dot')
nx.draw(G, pos, with_labels=True, arrows=True, node_size=500, node_color='lightblue', font_size=8)
plt.title("Roadmap for Preparing for the Final Stage of the All-Russian School Olympiad in Informatics")
plt.show()
```
