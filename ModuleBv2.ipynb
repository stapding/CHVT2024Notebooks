## 2.2 Тенденция развития олипиадного движения

Поскольку набор данных взят за 2021 год, нужно будет сделать рандомные года (например от 2021 до 2023) для нахождения трендов и тенденции развития олимпиадных движений. После этого обучим модель и определим её точность.

```python
import random

# Сначала обновим столбец 'YEAR' случайными значениями между 2021 и 2023 годами.
df['YEAR'] = df['YEAR'].apply(lambda x: random.randint(2021, 2023))

# Теперь подготовим данные для анализа с помощью машинного обучения, чтобы выявить тенденции и сделать прогнозы.
# Для этого будем использовать простую регрессионную модель, чтобы предсказать региональные тенденции участия и производительности с течением времени.

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Характеристики и целевая переменная (прогнозирование среднего балла SUM на основе региона, года и других характеристик)
X = df[['YEAR', 'REGION', 'GRADE_OF_COMPETITION', 'FINAL_SUM', 'FINAL_PERCENTAGE']]
X = pd.get_dummies(X, columns=['REGION'], drop_first=True)  # One-hot encoding for region
y = df['SUM']  # Target variable: total score (SUM)

# Разделение данных на обучающие и тестовые наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучение модели линейной регрессии
model = LinearRegression()
model.fit(X_train, y_train)

# Прогнозирование на тестовом наборе
y_pred = model.predict(X_test)

# Оценка производительности модели
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Теперь визуализируем временные тенденции, построив график среднего значения SUM за несколько лет для каждого региона
yearly_trends = df.groupby(['YEAR', 'REGION']).agg(
    avg_sum=('SUM', 'mean'),
    participants=('ID', 'count')
).reset_index()

# Построение трендов по годам
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_trends, x='YEAR', y='avg_sum', hue='REGION', marker='o')
plt.title('Average SUM Score Trend by Region Over Years')
plt.ylabel('Average SUM Score')
plt.show()

# Отображение точности модели и ее проницательности
model_accuracy = {'Mean Squared Error': mse, 'R-squared': r2}
model_accuracy
```

Для просмотра графика мешает легенда, уберём её и посмотрим на получившийся график.

```python
# Построение тех же трендов, но без легенды
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_trends, x='YEAR', y='avg_sum', hue='REGION', marker='o')
plt.title('Average SUM Score Trend by Region Over Years')
plt.ylabel('Average SUM Score')
plt.legend([], [], frameon=False)  # Remove the legend
plt.show()
```

Теперь можно что-то разобрать. Для определения региона можно воспользоваться легендой в предыдущем графике.

Машинное обучение выявило следующие результаты по анализу тенденций олимпиадного движения:

**1. Точность модели:**
- Среднеквадратичная ошибка (MSE): 200.31 — указывает на некоторое отклонение предсказанных значений от реальных данных.
- Коэффициент детерминации (R²): 0.13 — модель объясняет около 13% вариации данных, что говорит о наличии слабой, но существующей зависимости между характеристиками участников и их результатами.

**2. Тенденции по регионам:**
- Анализ временных трендов показал изменение среднего балла участников (SUM) по годам в разных регионах. В некоторых регионах наблюдаются положительные тенденции по увеличению баллов, в других — стабильность или снижение.

Эти данные указывают на слабую зависимость результатов олимпиадного движения от года и региона. 

# 3. Проверка нормальности распределения

```python
# Давайте сгруппируем данные по регионам и построим гистограммы для каждого региона по показателю SUM
regions = df['REGION'].unique()

# Настройка подплощадки для гистограмм
fig, axes = plt.subplots(len(regions)//4 + 1, 4, figsize=(15, 20))
axes = axes.flatten()

# Пройдёмся по каждому региону и постройм гистограмму оценок SUM
for i, region in enumerate(regions):
    region_data = df[df['REGION'] == region]
    axes[i].hist(region_data['SUM'], bins=10, color='skyblue', edgecolor='black')
    axes[i].set_title(f'{region}')
    axes[i].set_xlabel('SUM')
    axes[i].set_ylabel('Frequency')

# Настроим макет, чтобы избежать дублирования
plt.tight_layout()
plt.show()
```

Потчи ничего не видно, можно попробовать разделить регионы по группам, например по количеству участников.

```python
# Cначала сгруппируем регионы по количеству участников и разделим их на более мелкие группы для лучшей визуализации

# Сортировка регионов по количеству участников
region_participants = df.groupby('REGION')['ID'].count().sort_values(ascending=False)

# Разделение регионов на 3 группы по количеству участников
group_1 = region_participants[:len(region_participants)//3].index  # Top third
group_2 = region_participants[len(region_participants)//3:2*len(region_participants)//3].index  # Middle third
group_3 = region_participants[2*len(region_participants)//3:].index  # Bottom third

# Функция для построения гистограмм для заданной группы регионов
def plot_region_histograms(region_group, title):
    fig, axes = plt.subplots(len(region_group)//2 + 1, 2, figsize=(12, 15))
    axes = axes.flatten()

    for i, region in enumerate(region_group):
        region_data = df[df['REGION'] == region]
        axes[i].hist(region_data['SUM'], bins=10, color='skyblue', edgecolor='black')
        axes[i].set_title(f'{region}')
        axes[i].set_xlabel('SUM')
        axes[i].set_ylabel('Frequency')

    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()

# Построение гистограмм для каждой группы
plot_region_histograms(group_1, "Regions with Most Participants")
plot_region_histograms(group_2, "Regions with Medium Number of Participants")
plot_region_histograms(group_3, "Regions with Fewest Participants")
```

На основе гистограмм для различных групп регионов по количеству участников можно сделать следующие выводы:

### 1. **Регионы с наибольшим количеством участников**:

### 2. **Регионы с средним количеством участников**:

# 4. Математический анализ данных

Для выполнения математического анализа и сравнительного анализа между результатами в двух различных районах с использованием F-теста (критерий Фишера-Снедекора) или критерия Стьюдента (t-тест), мы можем сравнить дисперсии (для F-теста) или средние значения (для t-теста) результатов двух районов.
- F-тест: сравнение дисперсий двух выборок.
- t-тест Стьюдента: сравнение средних значений двух выборок.

Давайте начнем с выбора двух районов и проведем оба теста.

```python
from scipy import stats

# Сравним два региона, например, «Москва» и «Московская область», так как у них высокие показатели участия.
region_1 = df[df['REGION'] == 'Москва']['SUM']
region_2 = df[df['REGION'] == 'Московская область']['SUM']

# Выполнение F-теста (сравнение дисперсий)
f_stat, p_value_f = stats.levene(region_1, region_2)

# Выполнение Т-теста (сравнение средних)
t_stat, p_value_t = stats.ttest_ind(region_1, region_2, equal_var=False)  # Assuming unequal variance for the T-test

# Отображение результатов F-теста и T-теста
test_results = {
    'F-test (Variance Comparison)': {'F-statistic': f_stat, 'p-value': p_value_f},
    'T-test (Mean Comparison)': {'T-statistic': t_stat, 'p-value': p_value_t}
}

test_results
```

Результаты сравнительного анализа между результатами участников из Москвы и Московской области с помощью F-теста и T-теста:

1. **F-тест (для сравнения дисперсий)**:
   - **F-статистика**: 397.39
   - **p-значение**: 6.94e-85
   - Это крайне низкое p-значение говорит о том, что дисперсии результатов между Москвой и Московской областью значительно различаются. Разброс баллов в двух регионах не одинаков.

2. **T-тест (для сравнения средних значений)**:
   - **T-статистика**: 1.70
   - **p-значение**: 0.0899
   - Это p-значение больше стандартного порога значимости 0.05, что говорит о том, что средние результаты (SUM) между Москвой и Московской областью не различаются статистически значимо.

### Выводы:
- Дисперсии (разброс результатов) между Москвой и Московской областью сильно различаются, но средние значения баллов (SUM) не показывают значимых различий.
- Это может свидетельствовать о том, что в Москве разброс результатов более широкий, возможно, за счет большего числа участников и разных уровней подготовки.
